```python
import sys
import bisect

def eratosthenes(n):
    is_prime_arr = [True] * (n + 1)
    is_prime_arr[0] = is_prime_arr[1] = False
    for i in range(2, int(n**0.5)+1):
        if is_prime_arr[i]:
            for j in range(i*i, n+1, i):
                is_prime_arr[j] = False
    primes = [i for i, val in enumerate(is_prime_arr) if val]
    return primes

def min_remainder(primes, n):
    min_rem = n
    sqrt_n = int(n**0.5)
    idx = bisect.bisect_left(primes, sqrt_n)
    left = max(0, idx-10)
    right = min(len(primes)-1, idx+10)

    for i in range(left, right):
        p1, p2 = primes[i], primes[i+1]
        area = p1 * p2
        rem = n - area
        if rem >= 0 and rem < min_rem:
            min_rem = rem
    return min_rem

Q = int(sys.stdin.readline())
queries = [int(sys.stdin.readline()) for _ in range(Q)]

max_n = max(queries)
primes = eratosthenes(int(max_n**0.5)*2 + 10)

for n in queries:
    print(min_remainder(primes, n))
```
