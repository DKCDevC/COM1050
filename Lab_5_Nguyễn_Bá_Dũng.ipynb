```python
def is_prime(n):
    if n < 2:
        return False
    if n == 2 or n == 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True
```

```python
def is_perfect(n):
    if n < 2:
        return False
    sum_divisors = 0
    for i in range(1, n):
        if n % i == 0:
            sum_divisors += i
    return sum_divisors == n
```

```python
def calculator(num1, num2, operator):
  if operator == "+":
    return num1 + num2
  elif operator == "-":
    return num1 - num2
  elif operator == "/"
    return num1 / num2
  elif operator == "*":
    return num1 * num2
  else:
    return "Invalid operator"
```

```python
def factorial (n):
  if n <= 1:
    return 1
  return n * factorial(n-1)
```

```python
def cbd(list, n):
  for i in range(len(list)):
    if list[i] = n:
      return i
  return -1
```

```python
def sum_digits(n):
    s = 0
    while n > 0:
        s += n % 10
        n //= 10
    return s
```

```python
def is_isomorphic(a: str, b: str) -> bool:
    mapping = {}
    mapped_values = set()

    for char_a, char_b in zip(a, b):
        if char_a in mapping:
            if mapping[char_a] != char_b:
                return False
        else:
            if char_b in mapped_values:
                return False
            mapping[char_a] = char_b
            mapped_values.add(char_b)

    return True
```
