```python
def has_distinct_digits(n):
    mask = 0
    while n > 0:
        d = n % 10
        if mask & (1 << d):
            return False
        mask |= (1 << d)
        n //= 10
    return True

n = int(input("Nhập n: "))

result = []
i = 1

while i * i <= n:
    sq = i * i
    if has_distinct_digits(sq):
        result.append(sq)
    i += 1

if result:
    print(*result)
else:
    print("(no number)")
```

```python
def collatz_length(x):
    length = 1
    while x != 1:
        if x % 2 == 0:
            x //= 2
        else:
            x = 3*x + 1
        length += 1
    return length

while True:
    n = int(input("Nhập n (>0): "))
    if n > 0:
        break

max_len = 0
min_x = 1

for x in range(1, n+1):
    l = collatz_length(x)
    if l > max_len:
        max_len = l
        min_x = x
    elif l == max_len:
        min_x = min(min_x, x)

print(min_x, max_len)
```

```python
def sum_of_divisors(n):
    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i
    return total

a = int(input("Nhập số thứ nhất: "))
b = int(input("Nhập số thứ hai: "))

if sum_of_divisors(a) == b and sum_of_divisors(b) == a:
    print("true")
else:
    print("false")
```

```python
import numpy as np

total_animals = int(input("Nhập tổng số con: "))
total_legs = int(input("Nhập tổng số chân: "))

A = np.array([[1, 1],
              [2, 4]])
B = np.array([total_animals, total_legs])

try:
    solution = np.linalg.solve(A, B)
    g, c = solution

    if g >= 0 and c >= 0 and g.is_integer() and c.is_integer():
        print(int(g), int(c))
    else:
        print("invalid")

except np.linalg.LinAlgError:
    print("invalid")
```
